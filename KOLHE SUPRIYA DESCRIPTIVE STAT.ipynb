 ## Q.1 The average selling price of a food item in a restaurant is Rs 250.00 and the standard deviation 
of the price is Rs 47.00. The median value is an indication of the central value of the data and is 
close to the mean of Rs 202.00. If you know these details, will you become a regular customer 
of the restaurant?

# Import the necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Define the problem

```python
average_price = 250.00
standard_deviation = 47.00
median_value = 202.00
```

# Calculate the z-score

```python
z_score = (median_value - average_price) / standard_deviation
```

# Interpret the z-score

```python
from scipy.stats import norm

probability = norm.cdf(z_score)
```

# Provide the solution

```python
if probability > 0.5:
    print("Based on the given details, the median value is close to the mean, indicating a central value. Therefore, I would consider becoming a regular customer of the restaurant.")
else:
    print("Based on the given details, the median value is not close to the mean, indicating a deviation from the central value. Therefore, I would not become a regular customer of the restaurant.")
```

## Q.4 During a survey, the ages of 80 patients, infected by COVID and admitted to one of the city 
hospitals, were recorded. The collected data is represented in the less-than-cumulative 
frequency distribution table. Read the questions below the table and provide your answers.
Age (in yrs.) No. of Patients
5 - 15 6
15 - 25 11
25 - 35 21
35 - 45 23
45 - 55 14
55 - 65 5
a. Which class interval has the highest frequency?
b. Which age was affected the least?
c. How many patients of age 45 years and above were admitted?
d. Which is the modal class interval in the above dataset?
e. What is the median class interval of age?

## Import the necessary libraries

```python
# Import the necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the problem
frequency_table = [
    (6, (5, 15)),
    (11, (15, 25)),
    (21, (25, 35)),
    (23, (35, 45)),
    (14, (45, 55)),
    (5, (55, 65))
]

# Find the class interval with the highest frequency
highest_frequency = max(frequency_table, key=lambda x: x[0])
class_interval_highest_frequency = highest_frequency[1]

# Find the age affected the least
lowest_frequency = min(frequency_table, key=lambda x: x[0])
age_affected_least = lowest_frequency[1]

# Count the number of patients aged 45 years and above
patients_45_and_above = sum([frequency[0] for frequency in frequency_table if frequency[1][0] >= 45])

# Find the modal class interval
modal_class_interval = class_interval_highest_frequency

# Find the median class interval
cumulative_frequencies = np.cumsum([frequency[0] for frequency in frequency_table])
median_index = np.argmax(cumulative_frequencies >= 40)  # Assuming the median is at the 40th patient
median_class_interval = frequency_table[median_index][1]

# Output
print("a. Class interval with the highest frequency:", class_interval_highest_frequency)
print("b. Age affected the least:", age_affected_least)
print("c. Number of patients aged 45 years and above:", patients_45_and_above)
print("d. Modal class interval:", modal_class_interval)
print("e. Median class interval of age:", median_class_interval)
```

## Q.7 Calculate the z score of the following numbers: X = [4.5,6.2,7.3,9.1,10.4,11]

```python
import numpy as np

# Given list of numbers
X = [4.5, 6.2, 7.3, 9.1, 10.4, 11]

# Calculate mean and standard deviation
mean_X = np.mean(X)
std_dev_X = np.std(X)

# Calculate z-scores
z_scores = [(x - mean_X) / std_dev_X for x in X]

print("Z-scores for the numbers:", z_scores)
```

# Q.17 Give the IQR of all the variables which are quantitative and continuous.


```python
# Select quantitative and continuous variables
quantitative_continuous_variables = ['Age', 'Experience', 'Income', 'CCAvg', 'Mortgage']

# Calculate the IQR for each variable
for variable in quantitative_continuous_variables:
    # Calculate the first quartile (Q1)
    q1 = df[variable].quantile(0.25)
    
    # Calculate the third quartile (Q3)
    q3 = df[variable].quantile(0.75)
    
    # Calculate the interquartile range (IQR)
    iqr = q3 - q1
    
    # Display the results
    print(f"IQR for {variable}: {iqr:.2f}")
```
